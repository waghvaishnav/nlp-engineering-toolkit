# Stop Words in Spacy :

```python
import spacy
```

```python
from spacy.lang.en.stop_words import STOP_WORDS

print(STOP_WORDS)
```

# Identifying Stop Words in Spacy :

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("we just open our wings, the flying part is comming soon.")

print("stop words are :")
for token in doc:
    if token.is_stop:
        print(token)
```

# Stop Words Removal in Spacy :

creating a function for removing stop words

```python
def preprocess_stop_words(text):
    doc = nlp(text)

    no_stop_words = [token.text for token in doc if not token.is_stop and not token.is_punct]
    return no_stop_words
```

```python
preprocess_stop_words("my name is the vaishnav wagh.")
```

```python
preprocess_stop_words("python for machine learning and data science")
```
