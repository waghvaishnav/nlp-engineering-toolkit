## Language Processing Pipeline in NLP [English]

```python
import spacy

nlp = spacy.load("en_core_web_sm")

nlp.pipe_names
```

```python
nlp.pipeline
```

### POS / Lemma

```python
doc = nlp("captain America ate 100$ of samosa. when he said i can do this all day.")

for token in doc:
    print(token, "|",token.pos_,"| ",token.lemma_)
```

## Named Entity Recognition With "ner"

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 Billion.")

for ent in doc.ents:
    print(ent.text," | ",ent.label_,"| ",spacy.explain(ent.label_))
```

## displacy in Spacy (Visualization)

```python
from spacy import displacy

displacy.render(doc,style="ent")
```

## Blank Pipeline Customization

```python
nlp = spacy.blank("en")

doc = nlp("Tesla Inc is going to acquire twitter for $45 Billion.")

for ent in doc.ents:
    print(ent.text," | ",ent.label_," | ",spacy.explain(ent.label_))
```

#### In this spacy entity there is no print any statements because of the blank object to resolve this we can add custom processing in this

```python
source = spacy.load("en_core_web_sm")

nlp = spacy.blank("en")

nlp.add_pipe("ner",source = source)

doc = nlp("Tesla Inc is going to acquire twitter for $45 Billion.")

for ents in doc.ents:
    print(ents.text," | ",ents.label_," | ",spacy.explain(ents.label_))
```
