```python
import spacy

nlp = spacy.load("en_core_web_sm")

nlp.pipe_names
```

```python
nlp.pipe_labels["ner"]
```

## Identifying Entities :

```python
doc = nlp("elon musk is the owner of Tesla Inc.")

for ent in doc.ents:
    print(ent.text," | ",ent.label_," | ",spacy.explain(ent.label_))
```

## Visualization using displacy :

```python
from spacy import displacy

displacy.render(doc,style="ent")
```

## Customization of Entities using Span [Spacy]

```python
doc = nlp("Rays are very badly perform in S1_show in 1980")

for ent in doc.ents:
    print(ent.text," | ",ent.label_," | ",spacy.explain(ent.label_))
```

```python
type(doc[0])
```

```python
type(doc[0:4])
```

* in the above code there is a name "Rays" and he is Perform in a show name "S1_show". but there is no any entity identified by the spacy model.

* To Overcome this problem we have need to Set Customized Entities.

```python
from spacy.tokens import Span

s1 = Span(doc,0,1,label="PERSON")
s2 = Span(doc,6,7,label="ORG")

doc.set_ents([s1,s2],default = "unmodified")
```

```python
for ent in doc.ents:
    print(ent.text," | ",ent.label_," | ",spacy.explain(ent.label_))
```
