# Stemming [NLTK]

```python
import nltk
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["eating","eats","eat","ate","adjustable","rafting","meeting"]

for word in words:
    print(word,"-->",stemmer.stem(word))
```

# Lemmatization [Spacy]

```python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp("The children were running faster ate the were watching.")

for token in doc:
  print(token,"-->",token.lemma_,token.lemma)
```

# Lemmatization [NLTK] 

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

nltk.download("wordnet")
nltk.download("omw-1.4")


nlp = spacy.load("en_core_web_sm")

doc = nlp("The children were running faster ate the were watching.")

for token in doc:
    print(token,"-->",lemmatizer.lemmatize(token.text))
```

# Lemmatization Customization [Spacy]

```python
nlp = spacy.load("en_core_web_sm")

ar = nlp.get("attribute_ruler")

ar.add([[{"TEXT":"bro"}],[{"TEXT":"brah"}]],{"LEMMA":"Brother"})

doc = nlp("bro, you wanna go? brah, don't say no! i am exhausted.")

for token in doc:
    print(token,"-->",token.lemma_," | ",token.lemma)
```
