## Customization in Spacy

```python
nlp = spacy.blank("en")

doc = nlp("gimme double pizza extra large healthy pizza.")

for tokens in doc:
    print(tokens)
```

Without Customization spacy will gives two tokens for word "gimme".

```python
nlp = spacy.blank("en")

from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("gimme",
  [
    {ORTH:"gim"},
    {ORTH:"me"}
  ])

doc = nlp("gimme double pizza extra large healthy pizza.")

for tokens in doc:
    print(tokens)
```

Without Customization spacy will gives two tokens for word "gimme".
