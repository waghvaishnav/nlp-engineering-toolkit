```python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp("dr. the smallest and fastest English model. commonly used for etc. learning, prototyping, and basic NLP tasks.")
```

### Sentence Tokenization in spacy

```python
for sentence in doc.sents:
    print(sentence)
```

### Word Tokenization in Spacy

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```

# NLTK Code

importing nltk

```python
import nltk
nltk.download('punkt_tab')
```

### Sentence tokenizer in nltk

```python
from nltk.tokenize import sent_tokenize


sent_tokenize("It’s the smallest and fastest English model. commonly used for learning, prototyping, and basic NLP tasks.")
```

### Word Tokenizer in nltk

```python
from nltk.tokenize import word_tokenize

text = "It’s the smallest and fastest English model, commonly used for learning."

sentences = word_tokenize(text)

print(sentences)
```
